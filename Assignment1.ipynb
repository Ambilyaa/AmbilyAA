```python
vg=data[data['Classification']=='Iris-virginica']  # creating a new dataframe for Iris-virginica
v=vg.reset_index(drop=True)                        # resetting the index of the new dataframe
v.plot(kind='line',figsize=(15,6))                 #plotting the line graph for the new dataframe 
plt.show()
```

The line graph above shows the range of values of petal length(PL),petal width(PW),sepal length(SL) and sepal width(SW) of Iris-virginica plant samples taken.

```python
st=data[data['Classification']=='Iris-setosa'] # creating a new dataframe for Iris-setosa
st.plot(kind='line',figsize=(15,6))            #plotting the line graph for the new dataframe
plt.show()
```

The line graph above shows the range of values of petal length(PL),petal width(PW),sepal length(SL) and sepal width(SW) of Iris-setosa plant samples taken.

```python
vs=data[data['Classification']=='Iris-versicolor']  # creating a new dataframe for Iris-versicolor
s=vs.reset_index(drop=True)                         # resetting the index of the new dataframe
s.plot(kind='line',figsize=(15,6))                  #plotting the line graph for the new dataframe
plt.show()
```

The line graph above shows the range of values of petal length(PL),petal width(PW),sepal length(SL) and sepal width(SW) of Iris-versicolor plant samples taken.

```python
sns.boxplot(x='SW',data=data)        
plt.show()
```

The boxplot graph above shows the range of values of sepal width of the 3 categories.
 The values above 4 and below 2 are acting as outliers.

```python
sns.scatterplot(x='SL',y='SW',hue='Classification',data=data) 
plt.show()
```

The scatterplot above shows relation between the sepal lengths and sepal widths of the three different categories of plants.

```python
sns.pairplot(data,hue='Classification') 
plt.show()
```

The pairplot above shows the pairwise relationships between the petal lengths,petal widths,sepal lengths and sepal widths of the three different categories of plants given in the dataset. Through this we can summarize the data given in the table by a single figure.
