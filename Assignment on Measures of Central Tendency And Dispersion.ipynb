**_Q.3(A). Who scores the most on average for math, reading and writing based on gender?_**

```python
mean_gen=data.groupby('gender')[['math score','reading score',"writing score"]].mean() #finding column-wise mean  
mean_gen
```

```python
plt.figure(figsize=(15,6))                       
mean_gen.plot.bar()                                         #plotting bar graph 
plt.title("Gender-wise Mean of scores")
plt.show()
```

By finding the mean of all the given scores based on gender, male students score the most for Mathematics and female students score the most for reading and writing.

**_Q.3(B). Who scores the most on average for math, reading and writing based on test preparation course?_**

```python
mean_tp=data.groupby('test preparation course')[['math score','reading score',"writing score"]].mean()
mean_tp
```

```python
mean_tp.plot.bar()
plt.title('Mean of scores vs Test Preparation course')
plt.show()
```

By finding the mean of all the given scores based on test preparation course, those students who completed the test preparation course score the most for Mathematics, reading and writing.

**_Q.5. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this._**

```python
math_25pt=data['math score'].quantile(0.25)
```

```python
math_top=data[data['math score']>math_25pt]
```

```python
math_top=math_top.sort_values(by='math score',ascending=False)
```

```python
math_top.head(250)
```

Using the quantile function, calculated the 25th percentile of the math score. Then found the students whose math scores are above 25th percentile score.Then sorted the top math students based on their math score in descending order. Here,the total no: of students in the dataset is 1000 and its 25% is 250. Finally printed the top 25% of students based on their math score.
